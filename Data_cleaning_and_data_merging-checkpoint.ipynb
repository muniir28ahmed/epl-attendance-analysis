```python
import pandas as pd

weather = pd.read_csv("all_weather_data.csv")
stadium = pd.read_csv("Football_Stadiums.csv")
attendance_22_to_23 = pd.read_csv("Premier_League.csv")
```

### __`Inspect column names for inconsistencies`__

```python
display(weather.columns)
display(stadium.columns)
display(attendance_22_to_23.columns)
```

### __`Standardise names and parse dates as datetime objects `__

```python
attendance_22_to_23.columns = attendance_22_to_23.columns.str.strip().str.lower().str.replace(" ", "_")
stadium.columns = stadium.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(" ", "_")
weather.columns = weather.columns.str.strip().str.lower().str.replace(" ", "_").str.replace(" ", "_")
stadium['city'] = stadium['city'].str.lower().str.strip().str.replace(" ", "_")
stadium['stadium'] = stadium['stadium'].str.lower().str.strip().str.replace(" ", "_")
stadium['hometeams'] = stadium['hometeams'].str.lower().str.strip().str.replace(" ", "_")
stadium['country'] = stadium['country'].str.lower().str.strip().str.replace(" ", "_")
weather['location'] = weather['location'].str.lower().str.strip().str.replace(" ", "_")
attendance_22_to_23['stadium'] = attendance_22_to_23['stadium'].str.lower().str.strip().str.replace(" ", "_")
attendance_22_to_23['home_team'] = attendance_22_to_23['home_team'].str.lower().str.strip().str.replace(" ", "_")
attendance_22_to_23['away_team'] = attendance_22_to_23['away_team'].str.lower().str.strip().str.replace(" ", "_")
weather['date'] = pd.to_datetime(weather['date'],format="mixed", dayfirst=True)
attendance_22_to_23['date']= pd.to_datetime(attendance_22_to_23['date'],format="mixed", dayfirst=True)
attendance_22_to_23['clock'] = pd.to_datetime(attendance_22_to_23['clock'],format='mixed').dt.time
attendance_22_to_23['home_team_clean'] = (attendance_22_to_23['home_team'].str.strip().str.lower()).str.replace(" ", "_")
stadium['home_team_clean'] = (stadium['hometeams'].str.strip().str.lower()).str.replace(" ", "_")
```

```python
team_map = {
    'brighton_and_hove_albion': 'albion',
    'west_ham_united': 'west_ham',
    'wolverhampton_wanderers': 'wolverhampton',
    'bournemouth':'afc_bournemouth',
    'tottenham_hotspur': 'tottenham',
    'brentford': 'brentford_fc'
}
```

```python
attendance_22_to_23['home_team_clean'] = (attendance_22_to_23['home_team_clean'].replace(team_map))
```
